# üìä Centralized Model Analysis (MLPRegressor)

This notebook evaluates the performance of the centralized model used to predict daily temperature based on weather features.

The model was selected based on the lowest RMSE and saved as `baseline_model.pkl`.  
Here, we compare the predicted temperature values with the actual ones from the test set using data in `centralized_predictions.csv`.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load prediction results
df_pred = pd.read_csv("../logs/centralized_predictions.csv")  # or adjust path based on where you saved it
df_pred.head()
```

## üîç Predicted vs Actual Temperature

The following plot compares predicted temperature values from the centralized model with the actual test data.

```python
plt.figure(figsize=(10, 6))
plt.plot(df_pred['actual_temp'], label="Actual", linestyle='--', marker='o')
plt.plot(df_pred['predicted_temp'], label="Predicted", linestyle='-', marker='x')
plt.title("Predicted vs Actual Temperature (Centralized MLPRegressor)", fontsize=14)
plt.xlabel("Test Set Index (Days)", fontsize=12)
plt.ylabel("Temperature (¬∞C)", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üìè Model Performance Evaluation

We calculate common regression metrics to assess the model's accuracy:
- **MAE (Mean Absolute Error)**
- **RMSE (Root Mean Squared Error)**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt

mae = mean_absolute_error(df_pred["actual_temp"], df_pred["predicted_temp"])
rmse = sqrt(mean_squared_error(df_pred["actual_temp"], df_pred["predicted_temp"]))

print(f"‚úÖ MAE: {mae:.2f} ¬∞C")
print(f"‚úÖ RMSE: {rmse:.2f} ¬∞C")
```

## üìå Insights Summary

- The centralized MLPRegressor model demonstrates strong predictive performance, achieving a **MAE of 0.59¬∞C** and **RMSE of 0.72¬∞C** on the test set.
- The error values are relatively low, suggesting that the model is able to capture key temperature trends with high accuracy.
- Visual comparison between actual and predicted values shows tight alignment, with only minor deviations in certain days‚Äîindicating good generalization.
- This centralized baseline provides a solid benchmark for future comparisons with federated models.
- Potential next steps could include analyzing performance across seasons, incorporating time-series-aware models (e.g., LSTM), or simulating Non-IID data splits.
